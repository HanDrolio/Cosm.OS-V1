```
# Simple text-based adventure game: Shadows of the Keep

import random

# 1. Game State Variables:
game_name = "Shadows of the Keep"  # The name of the game
player_inventory = ["Fruit (various)", "Dusty map", "Indigo cloak"]  # List of items the player carries
current_location = "Royal Antechamber"  # String describing the player's current location

# 2. Core Game Functions:

def display_inventory():
    """
    Displays the player's current inventory.

    Returns:
        bool: True to indicate the game should continue.
    """
    if player_inventory:
        print("Inventory:", ", ".join(player_inventory))
    else:
        print("Inventory: Empty")
    return True

def roll_dice():
    """
    Simulates rolling a ten-sided dice.

    Returns:
        int: A random integer between 0 and 9 (inclusive).
    """
    return random.randint(0, 9)

def handle_action(action):
    """
    Handles player actions and updates the game state accordingly.

    Args:
        action (str): The action the player wants to perform.

    Returns:
        bool: True if the game should continue, False if the game should exit.
    """
    # Dictionary to map actions to functions
    actions = {
        "inventory": display_inventory,
        "rolldice": handle_rolldice,
        "eat fruit": handle_eat_fruit,
        "eat a fruit": handle_eat_fruit,
        "examine map": handle_examine_map,
        "examine dusty map": handle_examine_map,
        "map": handle_examine_map,
        "take cloak": handle_take_cloak,
        "take indigo cloak": handle_take_cloak,
        "cloak": handle_take_cloak,
        "drop cloak": handle_drop_cloak,
        "drop indigo cloak": handle_drop_cloak,
        "south": handle_move_south,
        "s": handle_move_south,
        "2": handle_move_south,
        "north": handle_move_north,
        "n": handle_move_north,
        "8": handle_move_north,
        "west": handle_move_west,
        "w": handle_move_west,
        "4": handle_move_west,
        "east": handle_move_east,
        "e": handle_move_east,
        "6": handle_move_east,
        "rest": handle_rest,
        "rest with cloak": handle_rest_with_cloak,
        "rest using cloak": handle_rest_with_cloak,
        "savegame": handle_savegame,
        "exit": handle_exit,
        "quit": handle_exit,
        "bye": handle_exit,
    }

    # Get the corresponding function for the action
    action_function = actions.get(action)

    # If the action is recognized, call the function
    if action_function:
        return action_function()
    else:
        print("I don't understand that command. Try 'inventory', 'rolldice', or a direction (north, south, east, west).")
        return True

# 3. Action Handler Functions:

def handle_rolldice():
    """
    Handles the 'rolldice' action.

    Returns:
        bool: True to indicate the game should continue.
    """
    roll = roll_dice()
    print(f"You rolled a {roll}!")
    # Logic for dice roll outcomes based on context
    if current_location == "Royal Antechamber":
        print("The dice roll echoes in the elegant chamber.")
    elif current_location == "Torch-lit Corridor":
        print("The torchlight flickers as the dice clatters on the stone floor.")
    elif current_location == "Dining Hall":
        print("The dice rolls across the polished wooden table.")
    elif current_location == "Storage Room":
        print("Dust motes dance in the air as the dice settles.")
    return True

def handle_eat_fruit():
    """
    Handles the 'eat fruit' action.

    Checks if the player has fruit and consumes it if available.

    Returns:
        bool: True to indicate the game should continue.
    """
    if "Fruit (various)" in player_inventory:
        print("You eat some fruit. It's refreshing.")
    else:
        print("You don't have any fruit.")
    return True

def handle_examine_map():
    """
    Handles the 'examine map' action.

    Checks if the player has a map and describes it if available.

    Returns:
        bool: True to indicate the game should continue.
    """
    if "Dusty map" in player_inventory:
        print("You examine the dusty map. It shows the castle layout with an 'X' in the west wing.")
    else:
        print("You don't have a map to examine.")
    return True

def handle_take_cloak():
    """
    Handles the 'take cloak' action.

    Adds the indigo cloak to the player's inventory if not already present.

    Returns:
        bool: True to indicate the game should continue.
    """
    global player_inventory
    if "Indigo cloak" not in player_inventory:
        print("You take the indigo cloak.")
        player_inventory.append("Indigo cloak")
    else:
        print("You already have the cloak.")
    return True

def handle_drop_cloak():
    """
    Handles the 'drop cloak' action.

    Removes the indigo cloak from the player's inventory if present.

    Returns:
        bool: True to indicate the game should continue.
    """
    global player_inventory
    if "Indigo cloak" in player_inventory:
        print("You drop the indigo cloak.")
        player_inventory.remove("Indigo cloak")
    else:
        print("You don't have the indigo cloak.")
    return True

def handle_move_south():
    """
    Handles moving south based on current location.

    Updates the current_location variable.

    Returns:
        bool: True to indicate the game should continue.
    """
    global current_location
    if current_location == "Royal Antechamber":
        current_location = "Torch-lit Corridor"
        print("You head south back into the torch-lit corridor.")
    elif current_location == "Dining Hall":
        current_location = "Torch-lit Corridor"
        print("You head south through the double doors back into the torch-lit corridor.")
    elif current_location == "Torch-lit Corridor":
        current_location = "Starting Room" # Placeholder - adjust as needed
        print("You head south.")
    else:
        print("You can't go south from here.")
    return True

def handle_move_north():
    """
    Handles moving north based on current location.

    Updates the current_location variable.

    Returns:
        bool: True to indicate the game should continue.
    """
    global current_location
    if current_location == "Torch-lit Corridor":
        current_location = "Dining Hall"
        print("You head north into the grand dining hall.")
    elif current_location == "Dining Hall":
        current_location = "Storage Room"
        print("You head north to the storage room.")
    elif current_location == "Royal Antechamber":
        print("There doesn't seem to be a way north from here.")
    else:
        print("You can't go north from here.")
    return True

def handle_move_west():
    """
    Handles moving west based on current location.

    Updates the current_location variable.

    Returns:
        bool: True to indicate the game should continue.
    """
    global current_location
    if current_location == "Torch-lit Corridor":
        print("You head west into a darker part of the castle.") # Placeholder
    elif current_location == "Royal Antechamber":
        current_location = "Torch-lit Corridor"
        print("You head west back into the torch-lit corridor.")
    else:
        print("You can't go west from here.")
    return True

def handle_move_east():
    """
    Handles moving east based on current location.

    Updates the current_location variable.

    Returns:
        bool: True to indicate the game should continue.
    """
    global current_location
    if current_location == "Torch-lit Corridor":
        current_location = "Royal Antechamber"
        print("You head east into a brightly lit antechamber.")
    elif current_location == "Dining Hall":
        print("There doesn't seem to be a way to go directly east.")
    else:
        print("You can't go east from here.")
    return True

def handle_rest():
    """
    Handles the 'rest' action.

    Simulates resting with a dice roll outcome.

    Returns:
        bool: True to indicate the game should continue.
    """
    roll = roll_dice()
    if roll > 5:
        print("You manage to get some rest.")
    else:
        print("You struggle to find a comfortable position.")
    return True

def handle_rest_with_cloak():
    """
    Handles the 'rest with cloak' action.

    Simulates resting with a dice roll, with a better chance of success if the cloak is in the inventory.

    Returns:
        bool: True to indicate the game should continue.
    """
    if "Indigo cloak" in player_inventory:
        roll = roll_dice()
        if roll > 3:
            print("The cloak provides some comfort, and you rest a bit better.")
        else:
            print("Even with the cloak, rest is elusive.")
    else:
        print("You don't have a cloak to rest with.")
    return True

def handle_savegame():
    """
    Handles the 'savegame' action.

    Prints a message indicating the game state is saved (in this session).

    Returns:
        bool: True to indicate the game should continue.
    """
    print("Game saved (in this session's state).")
    # In a real implementation, you'd write to a file
    return True

def handle_exit():
    """
    Handles the 'exit' action.

    Prints a goodbye message and indicates the game should end.

    Returns:
        bool: False to indicate the game should exit.
    """
    print("Thanks for playing Shadows of the Keep!")
    return False


# 4. Core Game Loop:
# This loop continues until the player decides to exit the game.
print(f"Welcome to {game_name}!")

while True:
    print("\n---")
    # Display current location and inventory
    print(f"You are in the {current_location}.")
    display_inventory()

    # Get user input
    action = input("What do you do? > ").lower().strip()

    # Handle the player's action
    if not handle_action(action):
        break
```
